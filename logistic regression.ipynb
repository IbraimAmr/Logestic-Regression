```python
df = df[df['titlef'].str.contains('Master|Officer|Mrs|Miss|Royalty')]
df
```

```python
df["Survived"]=df["Survived"].map({"Yes": 1, "No": 0})   #convert admitted variable into numerical
#df["Survived"]=df["Survived"].map({"Yes": 1, "No": 0}) 
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
x = df[['Pclass','SibSp','Age','Parch']]
y = df['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)  #splitting data with test size of 25%

logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing model’s performance
#print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

```python
sns.regplot(x='Parch',y='Survived',data=df)
#To predict if the passenger is survived or not (yes or no) we can use logistic regression.
#more than 0.5 survived , less than 0.5 didn't survived
```

```python
sns.regplot(x='Pclass',y='Survived',data=df)
```

```python

from sklearn.metrics import confusion_matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
""" 1.2e+0.2 is true positive as the model predicted that the passenger survived and it was right 
14 is false positive where the model predict that the passenger was survived and it was wrong 
48 False Negative is where the model predicted that the passenger was not survived and it was wrong
37 is true negative where the model predicted that the passenger was not survived and it was right  """
```

```python
"""
ROC and AUC are measures of performance to give a comprehensive evaluation of classification models
ROC (receiver operating characteristics) It’s plotted with two metrics against each other.
TPR (True Positive Rate or Recall) and FPR (False Positive Rate) where the former is on y-axis and the latter is on x-axis.
ROC computes TPR and FPR at various thresholds settings. Raising the classification threshold classifies more items as negative,
therefore decreasing both false Positives and true Positives, and vice versa. This means the two metrics are correlated
positively.
AUC (area under the curve) 
AUC tells how much our model, regardless of our chosen threshold, is able to distinguish between the two classes. 
The higher it is the better the model is. In a nutshell, AUC describes the degree of separability that our model makes.
It has a value between 1 and 0.""" 
```
